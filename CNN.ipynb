```python
import cv2
```

```python
import tensorflow as tf
```

```python
import os
image_name = os.listdir(r'E:\ML_Codes\all1\train')
```

```python
label = []
for i in image_name:
    if i. split(".")[0] == 'cat':
        label.append(0) 
    elif i. split(".")[0] == 'dog':
        label.append(1)
```

```python
from tqdm import tqdm
loc = r'E:\ML_Codes\all1\train'
features = []
for i in tqdm(os.listdir(loc)):
    x = os.path.join(loc,i)
    f = cv2.imread(x,0)
    rf = cv2.resize(f,(50,50))
    features.append(rf)
```

```python
import numpy as np
Y = np.array(label)
X = np.array(features)/255
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.20)
```

```python
model = tf.keras.models.Sequential()
#input layer
model.add(tf.keras.layers.Flatten())  #.Dense(14700)

#Hidden layer
model.add(tf.keras.layers.Dense(128, activation = 'relu'))
model.add(tf.keras.layers.Dense(64, activation = 'relu'))

#Output layer
model.add(tf.keras.layers.Dense(2, activation = 'softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=5)
```

```python
ypred = model.predict(xtest)
```

```python
#np.argmax(ypred[0])
y_predicted = []
for i in ypred:
    y_predicted.append(np.argmax(i))
```

```python
(y_predicted == ytest).sum() / len(xtest)
```

```python
clabel = []
for i in image_name:
    if i. split(".")[0] == 'cat':
        clabel.append([0,1]) 
    elif i. split(".")[0] == 'dog':
        clabel.append([1,0])
```

```python
#[0,0,1]
#[0,1,0]
#[1,0,0]
```

```python
Yc = clabel
```

```python
X = X.reshape(25000,50,50,1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Yc,test_size=0.20)
```

```python
#tflearn
import tflearn
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression
```

```python
conv = input_data(shape=[None, 50, 50, 1], name='input')
```

```python
conv = conv_2d(conv, 46, 5, activation='relu')
conv = max_pool_2d(conv, 5)

conv = conv_2d(conv, 44, 5, activation='relu')
conv = max_pool_2d(conv, 5)

conv = conv_2d(conv, 40, 5, activation='relu')
conv = max_pool_2d(conv, 5)

conv = conv_2d(conv, 36, 5, activation='relu')
conv = max_pool_2d(conv, 5)
```

```python
conv = fully_connected(conv, 1024, activation='relu')
conv = dropout(conv, 0.5)
```

```python
conv = fully_connected(conv, 2, activation='softmax')
```

```python
conv = regression(conv,
                  optimizer='adam',
                  learning_rate=0.0002, 
                  loss='categorical_crossentropy', 
                  name='targets')
```

```python
model = tflearn.DNN(conv)
```

```python
model.fit({'input': xtrain}, {'targets': ytrain}, 
          n_epoch=10, 
          validation_set=({'input': xtest}, {'targets': ytest}), 
          show_metric=True)
```
